```python
# Import Dependencies
import pandas as pd
from pathlib import Path
import numpy as np
```

```python
csv_path_laborhour = Path("../2.Resource/Maximo_download.csv")
labor_hour = pd.read_csv(csv_path_laborhour,  sep=",", encoding='cp1252')
labor_hour
```

```python
labor_hour = labor_hour[["route","location_id","work_date","mjuf_code","labor_hour"]]
labor_hour
```

```python
labor_hour = labor_hour.groupby(["route"]).sum()["labor_hour"]
labor_hour = pd.DataFrame(labor_hour)
labor_hour.round(2)
```

```python
labor_hour.reset_index
labor_hour = labor_hour[(labor_hour["labor_hour"] < 23360.67)]
labor_hour.round(0)
```
