```python
value=pi_flow_dec172023["Value"]
datetime=pi_flow_dec172023["DateTime"]

plt.bar(datetime,value,color="r",alpha=0.5,align="center")
plt.ylim((0, 12))
tick_loactions=[value for value in datetime]
plt.xticks(tick_loactions,datetime,rotation=90)

plt.title("Level Sensor 1995 Record")
plt.xlabel("Date Time")
plt.ylabel("Flow ft/s")
plt.grid(axis = "y", zorder=0)
plt.savefig("output_data/Top_10_Happiest_Countries.png", bbox_inches = "tight")
plt.show()
```
